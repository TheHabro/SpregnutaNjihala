```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from Reading import read_and_smooth, our_lord_and_saviour_fourier
```

```python
time1, red_ready, blue_ready = read_and_smooth("osnovni_period2.csv")
frekvencije = []

time = np.linspace(8, 30, 250)

p0 = 0.18, np.pi, 5, 0.02
popt, pcov = curve_fit(lambda t, A, omega, fi, B: A*np.cos(omega*t + fi) + B, time1, red_ready, p0)
print(popt[1], np.sqrt(pcov[1][1]))

A, omega, fi, B = popt

fig, ax = plt.subplots(figsize=[8,7])

plt.scatter(time1,  red_ready, color="silver")
plt.plot(time, A*np.cos(omega*time + fi) + B, color="blue")
plt.xlim(10, 20)
```

```python
four, freqs, PDS = our_lord_and_saviour_fourier(time1, red_ready)


fig, ax = plt.subplots(figsize=[8,7])

plt.plot(freqs, PDS, color = "red")
plt.xlim(0, 100)
plt.xlabel("f(Hz)")
plt.ylabel("Amplituda (arbitrary unit)")

sub_axes = plt.axes([.3, .3, .5, .5]) 
sub_axes.plot(freqs, PDS, color = "red") 

sub_axes.set_xlim(0.4, 0.7)
sub_axes.set_xlabel("f(Hz)")
sub_axes.set_ylabel("Amplituda (arbitrary unit)")

plt.savefig("OsnovniFTT.png")
```

```python
indeks = np.where(PDS == max(PDS))[0]
frekvencije.append(2 * np.pi * abs(freqs[indeks][0]))
```

```python
inverz_four = np.fft.ifft(four)

fig, ax = plt.subplots(figsize=[8,7])

plt.scatter(time1, inverz_four, color="silver", linewidth=3, label="mjereni podaci")
plt.plot(time, A*np.cos(omega*time + fi), color="indigo", label="prilagodba na podatke")
plt.plot(time1, red_ready, linestyle = '--', color="black", label="inverzni FTT")

plt.legend()

plt.xlabel("t(S)")
plt.ylabel("V(V)")

plt.xlim(10.5, 20)

plt.savefig("OsnovniKriv.png")
```
